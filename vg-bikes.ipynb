## 3. Given all the weather data here, find the average precipitation per month, and the median precipitation.

```python
df3.head(10)
def month(date):
    return int(date.split('/')[0])
df3['Month']=df3['Date'].apply(month)
df3.head(10)
```

```python
avg_precipitation=df3.groupby('Month')['Precipitation_In']
avg_precipitation.mean()
```

```python
avg_precipitation.median()
```

## 4. What’s the average number of bikes at a given bike station?

## 5. When a bike station is modified, is it more likely that it’ll lose bikes or gain bikes? How do you know?

```python
station=df2[['modification_date','install_dockcount','current_dockcount']].dropna()
station
```

```python
total = station['current_dockcount'].sum()-station['install_dockcount'].sum()
if total > 0:
    result = "gain bikes"
else:
    result = "lose bikes"

result
```
